#### Calculate speed using time and distance .. if distance is negative then break or continue

```python
choice="yes"  #initizliazation
while choice!="no":  #condition
    distance=float(input("plz enter distnace:"))  #2
    if distance<0: 
        print("Please enter positive distance")
        continue
    
    time=float(input("plz enter time:"))
    speed=distance/time
    print("The speed is:",speed)
    choice=input("Do you Want continue?")  #yes
```

# Quiz game using While loop 

```python
#questions
#answers
#while loop
#for loop
questions=["In which year did indiag gain independance?",
           "what is capital of india?",
           "which animal is known as the 'ship of the desert'?"
           ]

options=[["A.1974","B.1875","C.1947","D.1948"],
         ["A.Mumbai","B.Delhi","C.Kolkata","D.Pune"],
         ["A.Camel","B.Tiger","C.Cat","D.Dog"]]
Answers=["C","B","A"]
score=0
choice="yes"
while choice!="No":
    for i in range(len(questions)):  
        print(questions[i])
        for j in options[i]:  
            print(j)
        
        guess= input("Enter (A,B,C,D)")  
        if guess==Answers[i]:  
            print("Correct answer!")
            score=score+1
        else:
            print("wrong answer correct answer is",Answers[i])
            choice=input("Do you want to continue?")
            break
    
    if score==len(questions):  #3=3
        print("All answers are correct")
        choice=input("Do you want to continue?")
        
print("Quiz completed!")
print("Your score is:",score)
            
#outer for loop(3)==>(0,3)
#i=2
# j in options[1]==>"A.Mumbai","B.Delhi","C.Kolkata","D.Pune"
#A


#1 incorrect ==>stop
#3 correct ===>


            
            
        







           
        
```

```python
# calculate final price of product after applying a discount.
# discount depends on original price : 10% for prices below 50, 20% for above 50
price=float(input("Enter first price:"))  #0
l=[]
total=0
while price!=0:
    l.append(price)
    price=float(input("Enter next price:")) 
for i in l:
    total=total+i
    
if total>50:
    d=total*0.2
    pay=total-d
else:
    d=total*0.1
    pay=total
    
print(pay)
```
