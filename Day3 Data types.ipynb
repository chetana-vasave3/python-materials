# Solve the Quadratic equation

```python
#a*x*x+b*x+c=0
#64: +8,-8
a=1
b=5
c=6
import math
d=b*b-4*a*c
root1=(-b+math.sqrt(d))/2*a
root2=(-b-math.sqrt(d))/2*a
print(root1)
print(root2)
```

```python
a*root2*root2+b*root2+c
```
